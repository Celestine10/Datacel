```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv("Gaming_data.csv.csv")
df = df1.copy()
```

```python
X = df.iloc[:,0:1].values#Select just one column
y = df.iloc[:,1:2].values
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state=0)
dtr = dtr.fit(X,y)
```

```python
plt.scatter(X,y)
plt.plot(X, dtr.predict(X), color="red")
plt.title("Gaming Data (DECISION TREE REGRESSOR)")
plt.xlabel("Gaming Steps")
plt.ylabel("Points")
```

```python
dtr_pred = dtr.predict([[7.5]])
dtr_pred
```

```python
dtr_pred = dtr.predict([[8.5]])
dtr_pred
```

## Visualizing Decision Trees in Higher resolution

```python
X_grid = np.arange(min(X), max(X),0.01)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X,y)
plt.plot(X_grid, dtr.predict(X_grid), color="red")
plt.title("Gaming Data (DECISION TREE REGRESSOR)")
plt.xlabel("Gaming Steps")
plt.ylabel("Points")
```
