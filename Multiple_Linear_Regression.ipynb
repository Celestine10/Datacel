```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Org_data.csv")
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df1 = df.copy()
df1 = pd.get_dummies(df1, columns = ["State"])
```

```python
X = df1.drop(["Profit"],axis=1)
Y = df1["Profit"]
```

```python
#X = df1.iloc[:,:-1].values
#Y = df1.iloc[:,4].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr = lr.fit(X_train, Y_train)
```

```python
y_pred = lr.predict(X_test) 
```

```python
print(y_pred)
print("\n")
print(Y_test)
```

## Optimizing the Model

```python
import statsmodels.api as sm
```

```python
X = np.append(arr=np.ones((50,1)).astype(int), values = X, axis = 1)
```

```python
X_opt = X[:, [0,1,2,3,5]]
```

```python
lr_OLS = sm.OLS(endog=Y, exog=X_opt).fit()
```

```python
lr_OLS.summary()
```

```python
X_opt = X[:, [0,1,2,3,5]]
lr_OLS = sm.OLS(endog=Y, exog=X_opt).fit()
lr_OLS.summary()
```

```python
X_opt = X[:, [0,1,2,3]]
lr_OLS = sm.OLS(endog=Y, exog=X_opt).fit()
lr_OLS.summary()
```

```python
X_opt = X[:, [0,1,3]]
lr_OLS = sm.OLS(endog=Y, exog=X_opt).fit()
lr_OLS.summary()
```
